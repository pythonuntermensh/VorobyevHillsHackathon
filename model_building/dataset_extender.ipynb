```python
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv")

df.head()
```

```python
df = df.replace(r'^s*$', float('NaN'), regex = True)
df = df.dropna(subset=['Text'])
df = df.reset_index()
df = df.drop(['index'], axis=1)
```

```python
from ruwordnet import RuWordNet
import random
import nltk

nltk.download('popular')
#punkt
```

```python
wordnet = RuWordNet()

def get_synonyms(word):
    synonyms = []
    for syn in wordnet.get_synsets(word):
        synonyms.append(syn.title.lower())
    return synonyms

def replace_synonyms(text, n=1):
  sentences = []
  for j in range(n):
    words = nltk.word_tokenize(text)
    for i, word in enumerate(words):
        synonyms = get_synonyms(word)
        if synonyms:
            synonym = random.choice(synonyms)
            words[i] = synonym
    sentences.append(' '.join(words))
  return sentences
```

```python
import random
out_df = pd.DataFrame(columns=["Text", "Class", "r"])

for i in range(df.shape[0]):
  if i % 250 == 0:
    print(i)
  output_text = replace_synonyms(df.Text[i], n=3)
  for text in output_text:
    for j in range(len(text) // 20):
      pos = random.randint(0, len(text) - 1)
      code = ord(text[pos])
      code += random.randint(-1, 1)
      text = text[:pos] + chr(code) + text[pos+1:]
    out_df.loc[len(out_df)] = [text, df["Class"][i], ""]

out_df.head()

```

```python
out_df.to_csv("dataset.csv")
```
